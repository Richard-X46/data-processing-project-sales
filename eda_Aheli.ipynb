```python
genre_wise_sales = gameData.groupby('genre')['total_sales'].sum().sort_values(ascending=False)
genre_wise_sales.plot(kind='bar', figsize=(12,6))
plt.ylabel('Total Sales')
plt.title('Total Sales by Genre')
plt.tight_layout()
plt.show()
```

```python
console_wise_sales = gameData.groupby('console')['total_sales'].sum().loc[lambda x: x > 0].sort_values(ascending=True)
console_wise_sales.plot(kind='barh', figsize=(12,8))
plt.ylabel('Total Sales')
plt.title('Total Sales by Console')
plt.tight_layout()
# Increase the outer box size of the chart
plt.gcf().set_size_inches(16, 14)
# Add values on top of each bar
for idx, value in enumerate(console_wise_sales):
    plt.text(value + 5, idx, f'{value:.2f}', ha='left', va='center', fontsize=10)
plt.show()
```

```python
# Sales trends over time (by release_date)
gameData["release_date"] = pd.to_datetime(gameData["release_date"], errors="coerce")
year_wise_sales = gameData.groupby(gameData["release_date"].dt.year)["total_sales"].sum()
ax = year_wise_sales.plot(kind='bar', figsize=(12,6))
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')
ax.set_xticklabels([int(year) for year in year_wise_sales.index], rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Boxplots for numeric columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=gameData.select_dtypes(include=['number']), palette="Set2")
plt.title("Boxplots of Numeric Features", fontsize=16)
plt.show()
```

```python
def calculate_iqr_threshold(data, feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

#checking for outliers in 'price' and 'miles' columns
for feature in ['total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales']:
    lower_bound, upper_bound = calculate_iqr_threshold(gameData, feature)
    outliers = gameData[(gameData[feature] < lower_bound) | (gameData[feature] > upper_bound)]
    print(f"Outliers in {feature}: {outliers.shape[0]}")

outliers.head()
```
